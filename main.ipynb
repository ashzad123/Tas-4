```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataframe=pd.read_csv('10. HR-Employee-Attrition.csv')
```

```python
dataframe.head()
```

```python
dataframe.describe()
```

```python
department_counts = dataframe['Department'].value_counts()

# Create a bar plot
plt.figure(figsize=(10, 6))
department_counts.plot(kind='bar')
plt.title('Number of Employees in Each Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(dataframe['TotalWorkingYears'], dataframe['MonthlyIncome'], alpha=0.6)
plt.title('Relationship between Total Working Years and Monthly Income')
plt.xlabel('Total Working Years')
plt.ylabel('Monthly Income')
plt.grid(True)
plt.show()
```

```python
attrition_counts = dataframe[dataframe['Attrition'] == 'Yes']['JobRole'].value_counts()
total_counts = dataframe['JobRole'].value_counts()
attrition_rate = attrition_counts / total_counts * 100

# Create a bar plot
plt.figure(figsize=(12, 8))
attrition_rate.plot(kind='bar')
plt.title('Attrition Rate Among Different Job Roles')
plt.xlabel('Job Role')
plt.ylabel('Attrition Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate the average MonthlyIncome for male and female employees
average_income_by_gender = dataframe.groupby('Gender')['MonthlyIncome'].mean()

# Create a bar plot
plt.figure(figsize=(8, 6))
average_income_by_gender.plot(kind='bar', color=['blue', 'pink'])
plt.title('Average Monthly Income by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Monthly Income')
plt.xticks(rotation=0)
plt.show()
```

```python
import seaborn as sns
# Calculate the correlation matrix for Age and YearsWithCurrManager
correlation_matrix = dataframe[['Age', 'YearsWithCurrManager']].corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation between Age and Years With Current Manager')
plt.show()
```
